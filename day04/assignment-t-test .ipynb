```python
#Task 1
import numpy as np

# Generate random distributions
dist1 = np.random.randn(100) 
dist2 = np.random.randn(100) + 2
print("Distribution 1:", dist1)
print("Distribution 2:", dist2)
```

```python
#Task 2
from scipy.stats import ttest_ind

dist1 = np.random.randn(100)  
dist2 = np.random.randn(100) + 2

t_stat, p_value = ttest_ind(dist1, dist2)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
#Task 3
import matplotlib.pyplot as plt


dist1 = np.random.randn(100)  
dist2 = np.random.randn(100) + 2  

# Plot histograms
plt.hist(dist1, bins=10, alpha=0.8, label='Distribution 1')
plt.hist(dist2, bins=20, alpha=0.8, label='Distribution 1')
plt.legend()
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Task 4
# Create boxplots

dist1 = np.random.randn(100)  
dist2 = np.random.randn(100) + 2 

plt.boxplot([dist1, dist2], tick_labels=['Distribution 1', 'Distribution 2'])
plt.title("BOXPLOTS")
plt.ylabel("Values")
plt.show()
```

```python
# Task 4
import random

score = 0

for i in range(5):
    
    num_1 = random.randint(1, 500)  
    num_2 = random.randint(1, 800)  

    # Ask the user 
    answer = input(f"What is {num_1} - {num_2}? ")

    if int(answer) == num_1 - num_2:
        print("Correct!")
        score += 1  
    else:
        print(f"Wrong! The correct answer is {num_1 - num_2}.")

print(f"You scored {score}/5!")
```
